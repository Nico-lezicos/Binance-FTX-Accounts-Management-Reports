```python
#@title - Get Exchanges reports:
import os
import shutil
from datetime import datetime
currentDateTime = datetime.now().strftime("%y%m%d")

#Get Bin Concat
if os.path.exists(f'{bin_reports_path}Bin-Concat-{currentDateTime}.csv'):
    shutil.copy((f'{bin_reports_path}Bin-Concat-{currentDateTime}.csv'), (f'{ex1_reports_path}Bin-Concat-{currentDateTime}.csv'))
    print(f'Bin-Concat-{currentDateTime}.csv has been copied')
else:
    print("Can not copy the Binance file as it doesn't exists")


#Get FTX Concat
if os.path.exists(f'{ftx_reports_path}Ftx-Concat-{currentDateTime}.csv'):
    shutil.copy((f'{ftx_reports_path}Ftx-Concat-{currentDateTime}.csv'), (f'{ex2_reports_path}Ftx-Concat-{currentDateTime}.csv'))
    print(f'Ftx-Concat-{currentDateTime}.csv has been copied')
else:
    print("Can not copy the FTX file as it doesn't exists")


#!copy $ftx_reports_path + ftx*-{currentDateTime}.csv $ex2_reports_path
```

```python
#@title - Coins / Exchange :
import pandas as pd
from datetime import datetime
currentDateTime = datetime.now().strftime("%y%m%d")


# Binance Concat
df_binConcat = pd.read_csv(f'{ex1_reports_path}Bin-Concat-{currentDateTime}.csv', header=0, encoding='utf-8')
df_binConcat = df_binConcat.sort_values(by=['Stock-value'],ascending=False)
df_binConcat['exchangeName'] = 'Binance'
df_binConcat = df_binConcat[df_binConcat['Stock-value'] > 0.0]
#df_binConcat.reset_index(inplace=True)

# # FTX Concat
df_ftxConcat = pd.read_csv(f'{ex2_reports_path}Ftx-Concat-{currentDateTime}.csv', header=0, encoding='utf-8')
df_ftxConcat.rename(columns={'asset':'Symbol', 'totalAmount':'Quantity'}, inplace=True)
df_ftxConcat['exchangeName'] = 'FTX'
df_ftxConcat = df_ftxConcat.sort_values(by=['Quantity'],ascending=False)
df_ftxConcat = df_ftxConcat[df_ftxConcat['Stock-value'] > 0.0]
# df_ftxConcat.reset_index(inplace=True)

# # Exchange Concat
df_concat_exc = pd.concat([df_binConcat, df_ftxConcat])
df_concat_exc =  df_concat_exc.fillna(value={'Quantity':0, 'Stock-value':0, 'Stock-value':0})
df_concat_exc[['exchangeName', 'Symbol', 'Quantity', 'Price', 'Stock-value']].round({'Quantity':4, 'Price':4, 'Stock-value':2})
df_concat_exc = df_concat_exc.sort_values(by=['Stock-value'], ascending=False)

df_concat_exc[['exchangeName', 'Symbol', 'Quantity', 'Stock-value']].groupby(['Symbol']).sum(numeric_only=True).round({'Quantity':4, 'Price':4, 'coinSum':2,'Stock-value':2})
df_concat_exc['%'] = 100 * df_concat_exc['Stock-value'] / df_concat_exc['Stock-value'].sum()
df_concat_exc = df_concat_exc.sort_values(by=['Stock-value'],ascending=False).round({'Quantity':4, 'Price':4, 'Stock-value':2, '%':2})
#df_concat_exc.reset_index(inplace=True)


## Export Concat file
# check if file exist
if os.path.exists(f'{reports_path}Exc-Concat-{currentDateTime}.csv'):
    os.remove(f'{reports_path}Exc-Concat-{currentDateTime}.csv')
else:
    print("Can not delete the file as it doesn't exists")

# Export csv
df_concat_exc.round({'%':2})\
.to_csv(f'{reports_path}Exc-Concat-{currentDateTime}.csv'\
      , index = False, mode='a', header=True, encoding='utf-8')

print(f'export :')
!dir {reports_path}Exc-Concat-{currentDateTime}.csv

# #subflex_bin = df_concat_bin[['Symbol']]\

subflex_exc = df_concat_exc[['exchangeName', 'Symbol', 'Quantity', 'Price', 'Stock-value', '%']]\
.groupby(['exchangeName','Symbol']).sum(numeric_only=True)\
.round({'%':2, 'Quantity':4})
subflex_exc = subflex_exc.sort_values(by=['Stock-value'],ascending=False).round({'Quantity':2, 'Price':2, 'Stock-value':2, '%':2})

# # # Show full Dataframe
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
# pd.set_option('display.width', None)
# pd.set_option('display.max_colwidth', None)

#df_concat_exc
subflex_exc.round({'Quantity':2})
#df_binConcat.head()
#df_ftxConcat.head()
#df_concat_exc
```

```python
#@title - Coins / Simple Earn Flexible Balance Values :  > 1000
import matplotlib.pyplot as plt

df_condense = concat_exc.sort_values(by=['Stock-value'], ascending=False)
df_condense = df_condense[df_condense['Stock-value'] > 1000]
df_condense.reset_index(inplace=True) 

values = round(df_condense['Stock-value'],2).tolist()
bar_labels = df_condense['Symbol'].tolist()
coin = df_condense['Symbol'].tolist()


# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])

fig, ax = plt.subplots()

data_color = [x / max(values) for x in values]
bar_colors = plt.cm.rainbow(data_color)
# bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(coin, values, label=bar_labels, color=bar_colors)

ax.set_ylabel('Stock Value')
ax.set_title('Bin. & Ftx Coin Values')
# ax.legend(title='Coin')

# calling the function to add value labels
addlabels(coin, values)

plt.rcParams["figure.figsize"] = (30, 10)
plt.autoscale
plt.xlabel('values')


plt.show()
```
